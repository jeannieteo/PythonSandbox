```python
#pip install opencv-python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#This Python code reads an image using OpenCV, converts it from BGR to RGB color format and then displays it using Matplotlib.
src = cv2.imread(r'Mona_Lisa.jpg') #load an image as an array
```

```python
my_image  = cv2.cvtColor(src, cv2.COLOR_BGR2RGB) #converts image from BGR to RGB format for correct color display in Matplotlib.
# Display with Matplotlib
plt.imshow(my_image)
plt.title("RGB Image for Matplotlib")
plt.axis('off')
plt.show()
```

```python
print("oringinal image array shape: ", src.shape)
print("My_image array shape: ", my_image.shape)
```

```python
#show the values for pixel 100, 100
print("Pixel (100, 100) valules ", src[100][100] [:])
print("Pixel (100, 100) valules ", my_image[100][100] [:])
```

```python
my_image_resize = cv2.resize(my_image, (400, 600))
plt.imshow(my_image_resize)
plt.title("RGB Image resized for Matplotlib")
plt.axis('off')
plt.show()
```

```python
print("oringinal image array shape: ", src.shape)
print("My_image array shape: ", my_image_resize.shape)
```

```python
cv2.imwrite('resized_monalisa.jpg', my_image_resize)
```

```python
my_image_converted  = cv2.cvtColor(my_image_resize, cv2.COLOR_RGB2BGR) #converts it from RGB to BGR
cv2.imwrite('resized_monalisa_converted.jpg', my_image_converted)
```

```python
#convert to grayscale
my_image_gray = cv2.cvtColor(my_image_resize, cv2.COLOR_RGB2GRAY)
plt.imshow(my_image_gray, cmap = 'gray')
plt.title("RGB Image resized for Matplotlib")
plt.axis('off')
plt.show()
```

```python
# Thresholding is a very important operation in computer vision 
# it is a filtering operation that checks for pixel intensity up to a particular vaule.
# ANything below the value is removed, 
# we only get specific details like bright areas of the image
temp_image_gray = my_image_gray.copy()
#assert img is not None, "file could not be read, check with os.path.exists()"
ret,thresh1 = cv2.threshold(temp_image_gray,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(temp_image_gray,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(temp_image_gray,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(temp_image_gray,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(temp_image_gray,127,255,cv2.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [temp_image_gray, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    # subplots  (rows, cols, i)
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# Thresholding is a very important operation in computer vision 
# it is a filtering operation that checks for pixel intensity up to a particular vaule.
# ANything below the value is removed, 
# we only get specific details like bright areas of the image
drawing = cv2.imread('1280px-Codice_Atlantico_-_053v_detail.jpg')
drawing_gray = cv2.cvtColor(drawing, cv2.COLOR_RGB2GRAY)
temp_image_gray3 = drawing_gray.copy()
#assert img is not None, "file could not be read, check with os.path.exists()"
ret,thresh1 = cv2.threshold(temp_image_gray3,100,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(temp_image_gray3,100,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(temp_image_gray3,100,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(temp_image_gray3,100,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(temp_image_gray3,100,255,cv2.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [temp_image_gray3, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# averaging for bluring
temp_image_filter = my_image_resize.copy()
kernel = np.ones((3,3), np.float32)/9
print("# A 3x3 normalized box filter would look like below:")
print(kernel)
result = cv2.filter2D(temp_image_filter,-1, kernel)
result2 = cv2.filter2D(result, -1, kernel)
result3 = cv2.filter2D(result2, -1, kernel)
images = [result, result2, result3]
titles = ["first apply for smoothing","second apply blur","third apply"]
for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# This is done by convolving image with a normalized box filter. 
# It simply takes the average of all the pixels under kernel area and replace the central element. 
# This is done by the function cv2.blur() or cv2.boxFilter(). 
# Check the docs for more details about the kernel. 
# We should specify the width and height of kernel. 


blur = cv2.blur(temp_image_filter,(5,5))

plt.subplot(1,2,1),plt.imshow(temp_image_filter),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
kernel_sharpening = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
result4 = cv2.filter2D(temp_image_filter,-1, kernel_sharpening)
plt.subplot(1,2,1),plt.imshow(temp_image_filter),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(result4),plt.title('sharpening')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Edge detection: OpenCV provides several functions for performing edge detection, such as Canny(), Sobel(), and Laplacian(). 
# These functions can be used to identify edges in an image by looking for sharp changes in pixel intensity.

#img = cv2.imread('messi5.jpg', cv.IMREAD_GRAYSCALE)
#assert img is not None, "file could not be read, check with os.path.exists()"
edges = cv2.Canny(my_image_gray,100,255)
plt.subplot(121),plt.imshow(my_image_gray,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Load the pre-trained Haar cascade classifier for face detection
face_cascade = cv2.CascadeClassifier('haarcascade_profileface.xml')

faces = face_cascade.detectMultiScale(my_image_gray, scaleFactor=1.3, minNeighbors=5)

# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(my_image_gray, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Show the frame with detected faces
#plt.show(my_image_gray)
cv2.startWindowThread()
cv2.namedWindow("Image Window")
cv2.imshow('Face Detection', my_image_gray)
cv2.waitKey(0)
```
